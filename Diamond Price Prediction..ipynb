```python
def plotvariable(Diamond, variable):
    plt.figure(figsize=(16,4))
    #histogram
    plt.subplot(1,2,1)
    plt.hist(Diamond[variable], alpha=0.5)
    
    #boxplot
    plt.subplot(1,2,2)
    sns.boxplot(Diamond[variable])
    plt.show()
```

```python
plotvariable(Diamond, 'price')
```

### Detection of Outliers #through above box plot we can detect outliers 

```python
def find_boundaries(Diamond, variables):
    #Lets find the IQR (Inter quantile range)
    Q1 = Diamond[variables].quantile(.25)
    Q3 = Diamond[variables].quantile(.75)
    IQR= Q3-Q1
    lower_boundy = Q1 - 1.5*IQR
    upper_boundry = Q3 + 1.5*IQR
    
    return lower_boundy, upper_boundry
```

```python
lower_charges, upper_charges = find_boundaries(Diamond,'price')
print('Upper limit is ',upper_charges)
print('lower limit is ',lower_charges)
```

```python
#Adjusting outliers with upper limit 
#If the value is above upper limit charges there it will adjust to 34524.
#For example the charges was 40000 it will adjust it to 35000.

Diamond['price']= np.where(Diamond['price']> upper_charges,upper_charges,
                         np.where(Diamond['price']< lower_charges,lower_charges, Diamond['price']))
```

```python
plotvariable(Diamond, 'price')
```
